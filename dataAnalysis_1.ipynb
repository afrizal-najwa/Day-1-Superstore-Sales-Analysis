```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
try:
    df = pd.read_csv("data/SampleSuperstore.csv", encoding="utf-8")
except UnicodeDecodeError:
    df = pd.read_csv("data/SampleSuperstore.csv", encoding="latin-1")

print(df.head())
```

```python
# 2. Total penjualan (sales) & profit
total_sales = df["Sales"].sum()
total_profit = df["Profit"].sum()
print("\nTotal Sales:", total_sales)
print("Total Profit:", total_profit)
```

```python
# 3. Sales per kategori
sales_by_category = df.groupby("Category")["Sales"].sum().reset_index()
print("\nSales by Category:")
print(sales_by_category)
```

```python
# 4. Sub-kategori dengan sales tertinggi
top_subcat = df.groupby("Sub-Category")["Sales"].sum().reset_index().sort_values(by="Sales", ascending=False).head(5)
print("\nTop Sub-Categories:")
print(top_subcat)
```

```python
# 5. Region dengan profit terbesar
profit_by_region = df.groupby("Region")["Profit"].sum().reset_index().sort_values(by="Profit", ascending=False)
print("\nProfit by Region:")
print(profit_by_region)
```

```python
# 6. Visualisasi – Sales by Category
plt.figure(figsize=(6,4))
sns.barplot(x="Category", y="Sales", data=sales_by_category, palette="viridis")
plt.title("Total Sales by Category")
plt.show()

# 7. Visualisasi – Top 5 Sub-Categories
plt.figure(figsize=(8,4))
sns.barplot(x="Sub-Category", y="Sales", data=top_subcat, palette="magma")
plt.title("Top 5 Sub-Categories by Sales")
plt.show()
```
